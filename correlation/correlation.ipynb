```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df0=pd.read_excel('admission-pred-AnalysedData.xlsx',index_col=0,header=0)  # reading the .xlsx file and creating a data frame
df=pd.DataFrame(df0)
df
```

```python
df_corr=pd.DataFrame.corr(df,method='pearson',min_periods=1)
```

```python
df_corr  # correlation matrix 
```

```python
plt.matshow(df_corr)   # raw corellation graph
```

```python
#heatmap of the correlation of the data variables.
mask = np.zeros_like(df.corr(), dtype=np.bool)       # It returns array of zeros with the shape of df.corr
mask[np.triu_indices_from(mask)] = True              #Return the indices for the upper-triangle of an array.
plt.figure(figsize=(12,12))                          #Adjusted the figure size
sns.heatmap(df.corr(),mask=mask,annot=True,linewidths=1)   # This is used to plot rectangular data as a color-encoded matrix.
plt.show()
# As all the values are positive, we can conclude that all the variables are positively correlated to each other.
```
